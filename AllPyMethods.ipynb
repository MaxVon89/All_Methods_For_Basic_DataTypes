## Dictionary

```python
def dictMeth():
    i=0
    for method in dir(dict):
        i+=1
        print(f"#{i} {method}", sep="\n")

dictMeth()
```

```python
#Values()

bodyCount = {'Max':2, 'Ella': 5, 'Monica': 3, 'Nicole': 4}
print(sorted(bodyCount.values()))
```

```python
# keys()

print(bodyCount.keys())
```

```python
#pop()

Nic = bodyCount.pop('Nicole')
print(bodyCount)
# print(Nic)

```

```python
#popitem()
#Pops the last item

print(bodyCount.popitem())
```

```python
#copy()
# creates a shallow copy

bodyCount2 = bodyCount.copy()

print(bodyCount,bodyCount2)
print(bodyCount.popitem(),bodyCount2)
"""
Output:
('Monica', 3)
{'Max': 2, 'Ella': 5, 'Monica': 3}

dict. bodyCount is also affected since this is a shallow copy, even though they have diff. addresses (point to the same memory address).
"""
print(id(bodyCount))
print(id(bodyCount2))
```

```python
#get()
# gets key w/o raising an exception if not found

bodyCount.get('Max')
print(bodyCount.get('Jenny')) #not in bodyCount, returns None
print(bodyCount.get('Jenny', 'Not found')) #can give stand-in output if not found
```

```python
#setdef()
# same as get(), but if the val doesn't exist, it can set a val.

print(bodyCount.setdefault('Jenny', '8'))
print(bodyCount) #verify
```

```python
#clear()
# remmoves everything

bodyCount2.clear()
print(bodyCount2)
```

```python
#fromkeys()
# creates a dict with given keys (from an iterable, ex.: list) with 'None' vals

key1 = [1,2,3,4,5]

d2 = dict.fromkeys(key1)
print(d2)

d3 = dict.fromkeys(key1, 'e') #can set a def val, other than None
print(d3)
```

```python
# items()
# VERY important
# gives back an iterable of all the KV pairs

print(bodyCount.items())

for k,v in bodyCount.items():
    print(f" {k}'s body count is {v}")
```

```python
#update()
# add new KV pairs
#if keys are same, latest key will be retained

bodyCount.update({'Jenny':8})
print(bodyCount)


#union op

bodyCount |= {'Alexandra': 7, 'Eli': 9}
print(bodyCount)
```
