## 3.) Dictionary

```python
def dictMeth():
    i=0
    for method in dir(dict):
        i+=1
        print(f"#{i} {method}", sep="\n")

dictMeth()
```

```python
#Values()

birthDate = {'Max':2, 'Ella': 5, 'Monica': 3, 'Nicole': 4}
print(sorted(birthDate.values()))
```

```python
# keys()

print(birthDate.keys())
```

```python
#pop()

Nic = birthDate.pop('Nicole')
print(birthDate)
# print(Nic)

```

```python
#popitem()
#Pops the last item

print(birthDate.popitem())
```

```python
#copy()
# creates a shallow copy

birthDate2 = birthDate.copy()

print(birthDate,birthDate2)
print(birthDate.popitem(),birthDate2)
"""
Output:
('Monica', 3)
{'Max': 2, 'Ella': 5, 'Monica': 3}

dict. birthDate is also affected since this is a shallow copy, even though they have diff. addresses (point to the same memory address).
"""
print(id(birthDate))
print(id(birthDate2))
```

```python
#get()
# gets key w/o raising an exception if not found

birthDate.get('Max')
print(birthDate.get('Jenny')) #not in birthDate, returns None
print(birthDate.get('Jenny', 'Not found')) #can give stand-in output if not found
```

```python
#setdef()
# same as get(), but if the val doesn't exist, it can set a val.

print(birthDate.setdefault('Jenny', '8'))
print(birthDate) #verify
```

```python
#clear()
# remmoves everything

birthDate2.clear()
print(birthDate2)
```

```python
#fromkeys()
# creates a dict with given keys (from an iterable, ex.: list) with 'None' vals

key1 = [1,2,3,4,5]

d2 = dict.fromkeys(key1)
print(d2)

d3 = dict.fromkeys(key1, 'e') #can set a def val, other than None
print(d3)
```

```python
# items()
# VERY important
# gives back an iterable of all the KV pairs

print(birthDate.items())

for k,v in birthDate.items():
    print(f" {k}'s birth date is {v}")
```

```python
#update()
# add new KV pairs
#if keys are same, latest key will be retained

birthDate.update({'Jenny':8})
print(birthDate)


#union op

birthDate |= {'Alexandra': 7, 'Eli': 9}
print(birthDate)
```
